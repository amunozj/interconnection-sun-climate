```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import xarray as xr
```

```python
ds = xr.open_dataset('../data/toz_AERmon_GISS-E2-2-G_hist-sol_r1i1p3f1_gn_197601-200012.nc', decode_times=False)
ds0 = xr.open_dataset('../data/toz_AERmon_GISS-E2-2-G_historical_r1i1p3f1_gn_197601-200012.nc', decode_times=False)
ds['time'] = ds['time']/365.+1850
ds0['time'] = ds0['time']/365.+1850
```

```python
toz=ds['toz'].mean(dim='lon').sel(lat=0, method='nearest')
toz0=ds0['toz'].mean(dim='lon').sel(lat=0, method='nearest')
```

```python
toz.plot(figsize=(14,6))
toz0.plot()
```
